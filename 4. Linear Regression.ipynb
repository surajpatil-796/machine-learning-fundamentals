# Find the model that will return the best investment return

# Modeling with TV

```python
#In linear regression model it expects a 2 dimensional array input for x axis, therefore we write a axis columns in
#double square brakects
x_tv = df[["TV"]] 
x_radio = df[["radio"]]
x_news = df[["newspaper"]]
y = df["sales"]
```

```python
# spliting data for training and testing
x_train, x_test, y_train, y_test = train_test_split(x_tv, y, random_state=1, test_size=0.3)
#random state - if random state is not mentioned random data will be picked for every split
#test_size means the percentage of data to be kept for testing
```

```python
#create object of the LinearRegression
lr = LinearRegression()
```

```python
#fit method will find the slope(theta 1) and intercept (theta 0)
lr.fit(x_train,y_train)
```

```python
#return the value of intercept(theta 0)
lr.intercept_
```

```python
#returns the value of slope( theta 1)
lr.coef_
```

```python
#Testing 
#predict function will be predict y hat (yhat=theta0+theta1 * x)
y_pred = lr.predict(x_test)
x_test['TV'].to_list()
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
sns.scatterplot(data=df, x="TV", y="sales")
sns.lineplot(x_test['TV'].to_list(), y_pred)
plt.show()
```

# Modeling with Radio

```python
x_train, x_test, y_train, y_test = train_test_split(x_radio, y, random_state=1, test_size=0.3)
lr = LinearRegression()
#fit method will find the slope(theta 1) and intercept (theta 0)
lr.fit(x_train,y_train)
#return the value of intercept(theta 0)
lr.intercept_
#returns the value of slope( theta 1)
print(lr.coef_)
#predict function will be predict y hat (yhat=theta0+theta1 * x)
y_pred = lr.predict(x_test)
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2:",r2_score( y_test, y_pred))
sns.scatterplot(data=df, x="radio", y="sales")
sns.lineplot(x_test['radio'].to_list(), y_pred)
plt.show()
```

# Modeling with Newspaper

```python
x_train, x_test, y_train, y_test = train_test_split(x_news, y, random_state=1, test_size=0.3)
lr = LinearRegression()
#fit method will find the slope(theta 1) and intercept (theta 0)
lr.fit(x_train,y_train)
#return the value of intercept(theta 0)
lr.intercept_
#returns the value of slope( theta 1)
print(lr.coef_)
#predict function will be predict y hat (yhat=theta0+theta1 * x)
y_pred = lr.predict(x_test)
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2:",r2_score( y_test, y_pred))
sns.scatterplot(data=df, x="newspaper", y="sales")
sns.lineplot(x_test['newspaper'].to_list(), y_pred)
plt.show()
```

# Multiple Linear Regression

```python
### Computing with 3 parameters 
#     x1 = TV
#     x2 = radio
#     x3 = newspaper
x_trn = df [['TV', 'radio', 'newspaper']]
x_train, x_test, y_train, y_test = train_test_split(x_trn, y, random_state=1, test_size=0.3)
lr = LinearRegression()
lr.fit(x_train, y_train)
print(lr.intercept_)
print(lr.coef_)
y_hat = lr.predict(x_test)
print("MSE=",mean_squared_error(y_test, y_hat))
print("R2=", r2_score(y_test, y_hat))
```

```python
### Computing with 2 parameters 
#     x1 = TV
#     x2 = newspaper

x_tn = df[['TV', 'newspaper']]
x_train, x_test, y_train, y_test = train_test_split(x_tn, y, random_state=1, 
                                                    test_size=0.3)
lr = LinearRegression()
lr.fit(x_train, y_train)
y_hat = lr.predict(x_test)
print(mean_squared_error(y_test, y_hat))
print(r2_score(y_test, y_hat))
```

```python
### Computing with 2 parameters 
#     x1 = newspaper
#     x2 = radio
x_rn = df[['radio', 'newspaper']]
x_train, x_test, y_train, y_test = train_test_split(x_rn, y, random_state=1, 
                                                    test_size=0.3)
lr = LinearRegression()
lr.fit(x_train, y_train)
y_hat = lr.predict(x_test)
print(mean_squared_error(y_test, y_hat))
print(r2_score(y_test, y_hat))
```

```python
### Computing with 2 parameters 
#     x1 = TV
#     x2 = radio
x_tr = df[['TV', 'radio']]
x_train, x_test, y_train, y_test = train_test_split(x_tr, y, random_state=1, 
                                                    test_size=0.3)
lr = LinearRegression()
lr.fit(x_train, y_train)
y_hat = lr.predict(x_test)
print(mean_squared_error(y_test, y_hat))
print(r2_score(y_test, y_hat))
# We get best output with Multiple Linear Regression on indepent variables - TV and radio
```

# Assumption of Linear Regression
### Linear Regression will work best if all the assumption are satisfied

    1. Must have linear corelation - (PLease note if the data is not linear you will plot Polynomial regression)
    2. There should be little or no variance among residuals (Homoskedestic)
    3. There should be no multicolinearity (or any column that is dependent on the other column)
    4. Resudials must be normally distributed(bell curve)
    
##### You can use linear regression if Test 2, 3 or 4 fails but you can't use linear regression if Test 1 fails

```python
#Test 2: Check if data is homoskedestical
residuals = y_test - y_hat
sns.scatterplot(y_hat, residuals)
plt.show()
#if you observe the plot the data is hetroskedestical
# Test 2 fails
```

```python
# Test 3: There should be no multicolinearity (or any column that is dependent on the other column)
fig, axes = plt.subplots(1,3, figsize=(16, 6))
sns.scatterplot(data=df, x="TV", y="sales",ax=axes[0])
sns.scatterplot(data=df, x="radio", y="sales", ax=axes[1])
sns.scatterplot(data=df, x="radio", y="TV", ax=axes[2])
plt.show()
# Test 3 passes there is no multicolinearity in the all the 3 columns of TV, radio and sales
```

```python
# Test 4 Resudials must be normally distributed(bell curve)
sns.distplot(residuals)
plt.show()
# there is postive skewness in the output therefore Test 4 fails
```

# Polynomial Regression
When a straight line is unable to capture the patterns in the data.

```python
from sklearn.preprocessing import PolynomialFeatures
x = df[["TV","radio"]]
y = df["sales"]
```

```python
# Here the parameter passed in the PolynomialFeature is the degree
pf = PolynomialFeatures(2)
# Degree formula will be like
# y = TV x theta + TV * TV x theta + Radio x theta + Radio x radio x theta
x_poly = pf.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x_poly, y, random_state=1,test_size=0.3)
```

```python
lr= LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
# Check if data is homoskedestical
residuals = y_test - y_pred
sns.scatterplot(y_pred, residuals)
plt.show()
# Resudials must be normally distributed(bell curve)
sns.distplot(residuals)
plt.show()
```

```python
pf = PolynomialFeatures(3)
x_poly = pf.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x_poly, y, random_state=1,test_size=0.3)
lr= LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
# Check if data is homoskedestical
residuals = y_test - y_pred
sns.scatterplot(y_pred, residuals)
plt.show()
# Resudials must be normally distributed(bell curve)
sns.distplot(residuals)
plt.show()
```
