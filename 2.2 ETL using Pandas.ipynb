#### Group By

```python
matches = df.groupby('match_id')
```

```python
matches.groups
```

```python
# Top 3 run scorers
batsman_runs = df.groupby("batsman")["batsman_runs"].sum()
batsman_runs.head()
# but this data is not sorted
```

```python
# now we are working on sorting of data
sorted_runs = batsman_runs.sort_values(ascending=False)
sorted_runs.head(3)
```

```python
#Highest and lowest total score in an inning
match_runs = df.groupby(["match_id","inning"])["total_runs"].sum()
match_runs.head()
```

```python
# The reset_index() function is used to generate a new DataFrame or Series with the index reset.

# This is useful when the index needs to be treated as a column, or when the index is meaningless and 
# needs to be reset to the default before another operation
match_runs = match_runs.reset_index()
match_runs.head()
```

```python
match_runs["total_runs"].max()
```

```python
match_runs["total_runs"].min()
```

```python
# Average score of virat kohli
kohli = df[df["batsman"] == "V Kohli"]
```

```python
kohli_runs = kohli.groupby("match_id")["batsman_runs"].sum()
kohli_runs.head()
```

```python
kohli_runs = kohli_runs.reset_index()
kohli_runs.head()
```
