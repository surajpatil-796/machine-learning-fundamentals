```python
# Matplotlib is a visualization library in Python for 2D plots of arrays
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Line Plot
    Line plot is very basic matplotlib plot which is used to draw line chart.
    Line always contains two parameter x and y

```python
#visualization can possible with 1 dimension or 2 dimensional array
x = [1,3,5,7,9,11]
y = [11,14,15,19,24,25]
# In this the object name of the matplotlib will be visible
plt.plot(x,y)
```

```python
# to avoid the object name use show function
plt.figure(figsize=(7,7))
plt.plot(x,y)
plt.show()
```

```python
# 1 dimension plotting
plt.plot(x)
plt.show()
```

```python
# Creating wave line using sin wave in matplot
x = np.arange(0, 10, 0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

##### Understanding different attributes of matplotlib, which are commonly used across all graphs
    Figsize: to give size to chart area 
    Labels: to assign label to data (xlabel,ylabel)
    Title : to give title of chart 
    Limits: to start and end point to each axis (xlim, ylim)
    Ticks: used to rotate the characters visible on x and y axis (xticks, yticks) 
    Legend : gives label to plots 
    Color : used to give colors to plot (function parameter of plot)
Note: Multiple Plots: multiple line in one chart called multiple plot

```python
area = [600,800, 1000, 1200, 1500]
price = [15,18,25,35,40]
#increase size of the plot
#plt.figure(figsize=(width, height))
plt.figure(figsize=(7,7))
plt.plot(area,price)
#add information of the graph
#add title
plt.title("Area vs Price")
#add label for x and y
plt.xlabel("Area")
plt.ylabel("Price")
#adding limit to the plot
plt.xlim((400, 2000))
plt.ylim((5,50))
plt.show()
```

```python
plt.plot(area,price,'go-')
#add information of the graph
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
#rotate ticks
plt.xticks(rotation=90)
plt.yticks(rotation=45)
plt.show()
```

```python
#Multiple plots
x = np.arange(1,20)
y1 = np.sqrt(x)
y2 = np.log(x)
plt.figure(figsize=(6,6))
#changing the color of the lines
plt.plot(x,y1,color="yellow")
plt.plot(x,y2,color="red")
#legend in 2 line ploting
plt.legend(labels=['sqrt','log'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(x,x,color="blue")
plt.plot(x,y1,color="yellow")
plt.plot(x,y2,color="red")
plt.legend(labels=['linear','sqrt','log'])
plt.show()
```

### Histograms plots:
This is used for continuous data. Hist is used to define histogram type plot.

More attributes of matplotlib:
    
    axvline : gives verticle line on x axis
    bins:  define number of bins size required in plot (function parameter of hist)

```python
df = pd.read_csv("cars.csv")
df.info()
```

```python
df.describe()
```

```python
# before make graph write down your intution using describe functions. Is the data normally distrubutted?
plt.figure(figsize=(7,7))
plt.title("width frequency distribution")
plt.ylabel("freqeuncy")
plt.xlabel("width")
#bins are no of bars in histogram
plt.hist(df['width'],bins=10)
plt.show()
#graph is normally distributed
```

```python
# panda is made from numpy and matplotlib so all stuff we perform with matplotlib is possible with panda as well
df['width'].hist(bins=15)
plt.show()
```

```python
#show mean & median plot in graph
width_mean=df["width"].mean()
width_median=df["width"].median()
```

```python
plt.figure(figsize=(7,7))
plt.hist(df['width'],bins=10)
plt.axvline(width_mean,color="red")
plt.axvline(width_median,color="green")
plt.title("width frequency distribution")
plt.ylabel("freqeuncy")
plt.xlabel("width")
plt.show()
```

```python
#above same plot using pandas
plt.figure(figsize=(7,7))
df['width'].hist(bins=10)
plt.axvline(width_mean,color="red")
plt.axvline(width_median,color="green")
plt.title("width frequency distribution")
plt.ylabel("freqeuncy")
plt.xlabel("width")
plt.show()
```

### Bar plot
Majorly used for catogorial data
    
    Kind : type of plot specify bar in kind will give bar plot(bar, barh)

```python
#here the kind accepts bar this will create vertical bars
plt.figure(figsize=(9,9))
df["make"].value_counts().plot(kind="bar")
plt.title("Make vs Count")
plt.xlabel("Make")
plt.ylabel("Count")
plt.show()
```

```python
#here the kind accepts barh this will create horizontal bars
plt.figure(figsize=(9,9))
df["make"].value_counts().plot(kind="barh")
plt.title("Make vs Count")
plt.xlabel("Make")
plt.ylabel("Count")
plt.show()
```

### Pie chart
Used to define percentage based data. Pie function used to make pie plot 

    Autopct : used to give percentage distribution over data  
    Explode : used to extract pie out of plot 


```python
df["body-style"].value_counts()
```

```python
plt.figure(figsize=(7,7))
df["body-style"].value_counts().plot(kind="pie")
plt.show()
```

```python
#Pie using Pandas
plt.figure(figsize=(7,7))
df["body-style"].value_counts().plot.pie()
plt.show()
```

```python
#this will display the percentage on pie using pandas
plt.figure(figsize=(7,7))
df["body-style"].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
#this will display the percentage on pie using matplotlib
plt.figure(figsize=(7,7))
df["body-style"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.show()
```

```python
#this is to remove a part of sector away from the pie chart.
#with use of explode attribute it is possible
#the parameters in explode moves from top anticlockwise
# using pandas
plt.figure(figsize=(7,7))
df["body-style"].value_counts().plot.pie(autopct="%1.1f%%",
                                         explode=(0,0,0,0,0.2))
plt.show()
```

```python
#using matplotlib
plt.figure(figsize=(7,7))
df["body-style"].value_counts().plot(kind="pie", autopct="%1.2f%%", 
                                     explode=(0,0,0,0.1,0))
plt.show()
```
