```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cars.csv")
df.head()
```

### Line plot

```python
x = np.arange(10)
y = np.sqrt(x)
plt.figure(figsize=(7,7))
# we use lineplot function to plot line graph with sea born library
sns.lineplot(x,y)
plt.show()
```

### Distplot aka Histogram

```python
plt.figure(figsize=(7,7))
sns.distplot(df["width"], bins=10)
plt.show()
```

```python
width_mean = df["width"].mean()
width_median = df["width"].median()

plt.figure(figsize=(7,7))
sns.distplot(df["width"])

plt.axvline(width_mean, color="green")
plt.axvline(width_median, color="yellow")
plt.show()
```

###  Bar Plot / Count Plot

```python
#bar plot is called as count plot in seaborn
#count plot takes only one variable either x or y
plt.figure(figsize=(7,7))
sns.countplot(data=df, x="make")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df, x="make")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df, y="make")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df, x="body-style")
plt.show()
```

```python
sns.countplot(data=df, x="body-style", palette="Spectral_r")
plt.show()
```

```python
#use hue when you wanna distribute or group your data further
plt.figure(figsize=(7,7))
sns.countplot(data=df, x="body-style",hue="fuel-type")
plt.show()
```

### HeatMap
Heat Maps are graphical representations of data that utilize color-coded systems. The primary purpose of Heat Maps is to better visualize the volume of locations/events within a dataset and assist in directing viewers towards areas on data visualizations that matter most.

```python
df.corr()
```

```python
#light color means positively coerelated and dark color means negatively corelated
cor = df.corr()
plt.figure(figsize=(7,7))
sns.heatmap(cor)
plt.show()
```

```python
#annot will display all the numbers values of each cell in the heat map
plt.figure(figsize=(9,7))
sns.heatmap(cor,annot=True) 
plt.show()
```

```python
#cmap is used to change the color of the heat map
sns.heatmap(cor,annot=True, cmap="viridis") 
plt.show()
```

### Scatter Plot

```python
plt.figure(figsize=(7,7))
sns.scatterplot(data=df,x="width",y="price") #always keep the target on y axis
plt.show()
```

```python
plt.figure(figsize=(7,7))
#dot's colors will change as per hue
sns.scatterplot(data=df,x="width",y="price",hue="body-style") 
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=df,x="width",y="price",hue="make") 
plt.show()
```

### Box plot
Displays Q1, Q2 and Q3, majorly used to find outlier in data
    
    The box represents Q1, Q2 and Q3
    Lines represents wishker's 
    The dots outside the lines are outliers

```python
plt.figure(figsize=(10,4))
#box plot can be done either one column or two columns
sns.boxplot(data=df, x="price") 
plt.show()
```

```python
plt.figure(figsize=(10,4))
#if you wish to increase the range add whis parameter, by default it is 1.5
sns.boxplot(data=df, x="price", whis=3) 
plt.show()
```

```python
# box plot for multiple columns
plt.figure(figsize=(10,10))
sns.boxplot(data=df, x="price", y="make") 
plt.show()
```

### Sub Plots
Making multiple plots together

```python
x = [1,2,3,4]
y = [3,4,5,6]

#subplot accepts number of rows and columns
fig,axes = plt.subplots(2,2,figsize=(8,6))

#axes symbolizes the matrix cordinates
axes[0,0].plot(x,y)
axes[0,1].scatter(x,y)
axes[1,0].plot(x,y)
axes[1,1].scatter(x,y)

plt.show()
```

```python
#sub plot using sea born
fig,axes = plt.subplots(2,2,figsize=(16,10))
#you will pass an attribute ax to pass axes
sns.distplot(df["price"], ax=axes[0,0])
sns.countplot(data=df, y="make",ax=axes[0,1])
sns.scatterplot(data=df,x="width",y="price",hue="body-style",ax=axes[1,0])
sns.boxplot(data=df, x="price",ax=axes[1,1])
plt.show()
```
